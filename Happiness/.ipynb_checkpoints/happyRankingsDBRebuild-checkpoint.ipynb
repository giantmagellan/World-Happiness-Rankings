```python
# Function to assign csv files to a single array
def getCSVFiles():
    # Calls ALL csv files from the data path
    csvFiles = glob.glob(os.path.join(path, '*.csv'))

    # Empty array to assign the csv files
    dataframes = []

    # Assigns dataframes to empty array
    for csv in csvFiles:
        df = pd.read_csv(csv)
        dataframes.append(df)
    
    print(dataframes)
```

```python
# Function to create csv names
def createCSVNames():
    # Empty array to hold the new file names
    csvNames = []
    
    # Years of Rankings data
    years = ['2015', '2016', '2017', '2018', '2019']
    
    # Loop through desired years to create new names for each csv data file
    for year in years:
        csvName = 'rankings' + year
        csvNames.append(csvName)
        
    print(csvNames)
```
